```python
from sklearn.preprocessing import LabelEncoder # for label encoding  categorical data/ converting into numerical form. 
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(data_cat["ocean_proximity"])
```

```python
for col in data_cat:
    le = LabelEncoder()
    data_cat[col] = le.fit_transform(data_cat[col])
```

```python
data_new = pd.concat([data_num,data_cat],axis=1)# merging the categorical( now numeric) and numerical data.
```

```python
columns_to_check = [ 'longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 
                    'households', 'median_income', 'median_house_value', 'ocean_proximity']
```

```python
from scipy.stats import zscore # Z- score for outlier 
data_new_zscore = data_new[columns_to_check].apply(zscore)
threshold = 3
```

```python
outliers = (np.abs(data_new_zscore) > threshold).any(axis=1)
```

```python
New_data = data_new[~outliers] 
```

```python
New_data.to_csv('New_data.csv', index=False)# new data set after removing outliers 
```

```python
print(f"Outliers removed: {outliers.sum()} rows")
```

```python
New_data
```

```python
from sklearn.model_selection import train_test_split# importing train_test_split for model evaluation of subset set of data

x = New_data.drop(['median_house_value'], axis=1)
y = New_data['median_house_value']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2)# splitting data for train and test purpose.
```

```python
x_train 
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression # Importing LinearRegression for fitting a linear model for prediction. 
clf = LinearRegression() # creating a linear regression model object
```

```python
clf.fit(x_train, y_train) # trains the x and y
```

```python
clf.predict(x_test) # generate prediction based on the trained model.
```

```python
y_test
```

```python
clf.score(x_test, y_test)# model accuracy of the test data
```

```python
from sklearn.ensemble import RandomForestRegressor # RandomForestRegressor is for averaging to imporve the predictive accuracy and control over-fitting
forest = RandomForestRegressor()
forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
## we found randomforestregressor with the accuracy of 0.7947 (79.47%) is better than linear regression with accuracy of 0.6169(61.69%).
```
